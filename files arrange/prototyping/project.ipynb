```python
import shutil
from pathlib import Path

def main():
    print("--- File Organizer 3000 ---")
    
    # 1. Get validated inputs
    user_dir = get_valid_directory()
    target_folder_name = input("Enter name for the new organized folder: ").strip()
    
    # 2. Define the destination root
    destination_root = user_dir / target_folder_name
    
    # 3. Create the destination folder (if it doesn't exist)
    # exist_ok=True means it won't crash if the folder is already there
    destination_root.mkdir(parents=True, exist_ok=True)

    print("\nProcess Complete!")


def get_valid_directory():
    """Loops until the user provides a valid, existing directory."""
    while True:
        path_str = input("Insert the source directory path: ").strip()
        path_obj = Path(path_str)
        
        if not path_obj.exists():
            print("Error: That path does not exist.")
        elif not path_obj.is_dir():
            print("Error: That input is a file, not a folder.")
        else:
            return path_obj


def organize_files(source_dir, dest_root):
    """Iterates through source and copies files to destination based on extension."""
    
    # We list all items first to avoid the 'infinite loop' of copying 
    # a file that we just pasted.
    all_items = list(source_dir.iterdir())

    for item in all_items:
        # SKIP the destination folder itself so we don't try to copy it inside itself
        if item == dest_root:
            continue

        if item.is_file():
            # Get extension (e.g., '.txt') and remove the dot ('txt')
            # If no extension (empty string), we categorize as 'Other'
            extension = item.suffix.lower().lstrip('.') or "Other"
            
            # Create the specific subfolder (e.g., source/Organized/png)
            target_folder = dest_root / extension
            target_folder.mkdir(exist_ok=True)
            
            copy_item(item, target_folder)

        elif item.is_dir():
            # Handle folders
            target_folder = dest_root / "Folders" / item.name
            # We use copytree for folders
            try:
                shutil.copytree(item, target_folder, dirs_exist_ok=True)
                print(f"✅ Folder copied: {item.name} -> {target_folder.name}")
            except Exception as e:
                print(f"❌ Error copying folder {item.name}: {e}")


def copy_item(source_item, target_folder):
    """
    A single function to handle copying.
    It combines your previous 2 file-copying functions into one.
    """
    try:
        shutil.copy2(source_item, target_folder)
        print(f"✅ File copied: {source_item.name} -> {target_folder.name}/")
    except Exception as e:
        print(f"❌ Error copying {source_item.name}: {e}")


if __name__ == "__main__":
    main()
```

### 4\. Key Lessons from the Changes

**1. Handling Extensions the Pythonic Way**

  * **Your way:** `get_extention.split(".")[1]`
  * **My way:** `item.suffix.lower().lstrip('.')`
      * `item.suffix`: Gets the extension automatically (e.g., `.JPG`).
      * `.lower()`: Makes `.JPG` and `.jpg` go into the same folder (very important for Windows users\!).
      * `.lstrip('.')`: Removes the dot nicely.

**2. Boolean Logic for "Other"**

  * I used this trick: `extension = item.suffix ... or "Other"`.
  * In Python, empty strings evaluate to `False`. So if a file has no extension, the code automatically picks "Other". This removed an entire `if/else` block from your code\!

**3. The `if __name__ == "__main__":` block**

  * At the bottom, I added this check. This is a best practice. It prevents your `main()` function from running automatically if you try to import this script into another program later.
