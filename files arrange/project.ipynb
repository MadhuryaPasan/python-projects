```python
import shutil
from pathlib import Path
```

```python
home_dir = Path.home()

new_dir = home_dir / "Downloads"
```

```python
# get all available extentions
extentions = []
for file in new_dir.iterdir():
    # print(file)
    withdot = file.suffix
    if withdot:
        withoutdot = withdot.split(".")[1]
        if withoutdot not in extentions:
            extentions.append(withoutdot)
        
extentions
```

```python
main_destination = new_dir / "arranged"
print(main_destination)
```

```python
# create folders
for extention in extentions:
    sub_folders = main_destination/extention
    print(sub_folders)
    sub_folders.mkdir(parents=True, exist_ok=True)
```
