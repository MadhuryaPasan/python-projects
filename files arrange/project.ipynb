```python
from genericpath import exists
import shutil
from pathlib import Path


def main():
    # get user input dir that want to arranged
    user_dir = get_user_dir()
    # get user input for the folder that they want to coppy the files
    new_folder_name = get_new_folder_name()
    # create root folder to past the coppied files
    new_folder_dir = new_folder_for_coppied_files(user_dir, new_folder_name)
    arrange_files(user_dir, new_folder_dir)


def get_user_dir():
    """
    get dir from users and return the dir
    """
    while True:
        user_input_str = input("Insert the dir :")
        # conver the string
        user_dir = Path(user_input_str)
        # some vaidations
        if user_dir.exists():
            print("folder exists!")
            if user_dir.is_dir():
                print("it is a folder")
                break
            else:
                print("please check is it a folder")
        else:
            print("please check if the folder if exists")
    return user_dir


def get_new_folder_name():
    user_input = input("Enter new folder name")
    return user_input


def new_folder_for_coppied_files(root_dir, new_folder_name):
    """
    insert folder name that you need to coppy the aranged files.

    eg.->
        "new_folder"
    for sub dir->
        "new_folder/sub_folder"

    returns the created folder dir
    """
    try:
        # combine the root dir and new folder
        user_dir = Path(root_dir / new_folder_name)
        # create the folder in the root dir
        user_dir.mkdir(parents=True, exist_ok=True)
    except Exception as e:
        print("Error: Something went wrong when creating new folder!")
        print(e)
    else:
        print("Success: Folder created")
        return user_dir


def create_sub_folders_for_extentions(new_folder_dir, extention):
    """
    combine new folder dir with extention name to create new sub folder

    extention -> extention name
    new_folder_dir -> dir of the new folder
    """
    try:
        sub_folder_dir = Path(new_folder_dir / extention)
        # create folder
        sub_folder_dir.mkdir(parents=True, exist_ok=True)
    except Exception as e:
        print(
            "Error: Something went wrong when creating sub folders for the extentions"
        )
        print(e)
    else:
        print("Success: sub folder for ", extention, " created")


def coppy_file_with_extention(source_file, new_folder_dir, extention):
    try:
        extention_folder_dir = Path(new_folder_dir / extention)
        source_file = Path(source_file)
        # coppy file
        shutil.copy2(source_file, extention_folder_dir)
    except Exception as e:
        print("Error: Some thing went wrong when copping the file that has extention")
        print(e)
    else:
        print("Success: ", source_file.name, " coppied successfully")


def coppy_file_without_extention(source_file, new_folder_dir, subfolder_name):
    try:
        subfolder_dir = Path(new_folder_dir / subfolder_name)
        source_file = Path(source_file)
        # coppy file
        shutil.copy2(source_file, subfolder_dir)
    except Exception as e:
        print(
            "Error: Some thing went wrong when copping the file that did not has extention"
        )
        print(e)
    else:
        print("Success: ", source_file.name, " coppied successfully")


def coppy_folder(source_file, new_folder_dir, subfolder_name):
    try:
        source_file = Path(source_file)
        subfolder_dir = Path(new_folder_dir / subfolder_name / source_file.name)
        ignore_name = subfolder_name

        # ignore_func = shutil.ignore_patterns(new_folder_dir)
        def ignore_copied_dir(directory, contents):
            if ignore_name in contents:
                return [ignore_name]
            return []

        # coppy folder
        shutil.copytree(
            source_file, subfolder_dir, dirs_exist_ok=True, ignore=ignore_copied_dir
        )
        print(">>>>>", subfolder_dir)
        print(">>>>>", source_file)
    except Exception as e:
        print("Error: Some thing went wrong when copping the folder")
        print(e)
    else:
        print("Success: ", source_file.name, " coppied successfully")


dir_list = []


def arrange_files(root_dir, new_folder_dir):
    extentions = []
    # to privent the arranged file copping again and again
    for file in root_dir.iterdir():
        if new_folder_dir != file:
            dir_list.append(file)

    for file in dir_list:
        file = Path(file)
        # print(file.name)
        get_extention = file.suffix
        # remove the dot(.mp3 -> mp3)
        if get_extention:  # if a file has an extention.
            clean_extention_name = get_extention.split(".")[1]
            # create sub folders for extentions
            if clean_extention_name not in extentions:
                extentions.append(clean_extention_name)
                create_sub_folders_for_extentions(new_folder_dir, clean_extention_name)
            # coppy files
            coppy_file_with_extention(file, new_folder_dir, clean_extention_name)
        else:
            # create sub folder for the files/ folder that dose not have an extention
            other = "Other"
            other_dir = Path(new_folder_dir / other)
            if other_dir.exists():
                print("Werning: ", other, " folder alredy exist")
            else:
                create_sub_folders_for_extentions(new_folder_dir, other)
            if file.is_file():
                print("this is a file")
                coppy_file_without_extention(file, new_folder_dir, other)
            else:
                coppy_folder(file, new_folder_dir, other)


main()
```
